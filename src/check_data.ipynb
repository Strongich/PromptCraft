```python
import re

# Function to convert duration to minutes
def convert_duration_to_minutes(duration):
    duration = duration.strip()  # Remove leading and trailing whitespaces
    if 'h' in duration and 'm' in duration:
        # If duration is in the format "Xh Ym"
        hours, minutes = map(int, re.findall(r'\d+', duration))
        return hours * 60 + minutes
    elif 'h' in duration:
        # If duration is in the format "Xh"
        return int(re.search(r'\d+', duration).group()) * 60
    elif 'm' in duration:
        # If duration is in the format "Ym"
        return int(re.search(r'\d+', duration).group())
    else:
        # If duration has unexpected format
        return None

# Assuming df is your DataFrame and "Duration" is the column
df["Duration"] = df["Duration"].apply(convert_duration_to_minutes)
```
